```python
import pandas as pd
```

```python
dataset=pd.read_csv('dataset.csv')
dataset.head(3)
```

```python
dataset.iloc[:,10:]
```

```python
dataset.drop(columns=['PolicyNumber','RepNumber','Year'],inplace=True)
```

```python
dataset.columns.tolist()
```

```python
dataset.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Drop the single corrupted row containing "0" in claim date fields
dataset = dataset[dataset['DayOfWeekClaimed'] != '0']

# Separate features (x) and target (y)
x = dataset.drop(columns=['FraudFound_P'])
y = dataset['FraudFound_P']

# Perform a stratified train-test split to preserve fraud ratio
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# Reset index to avoid alignment issues during preprocessing joins
x_train = x_train.reset_index(drop=True)
x_test = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

print("Train shape:", x_train.shape)
print("Test shape:", x_test.shape)
```

```python
# Impute missing Age (0 values) to 16 (corresponding to '16 to 17' category)
x_train.loc[x_train['Age'] == 0, 'Age'] = 16
x_test.loc[x_test['Age'] == 0, 'Age'] = 16

# Map binary columns to 0/1
binary_cols = {
    'Sex': {'Male': 1, 'Female': 0},
    'AccidentArea': {'Urban': 1, 'Rural': 0},
    'Fault': {'Policy Holder': 1, 'Third Party': 0},
    'PoliceReportFiled': {'Yes': 1, 'No': 0},
    'WitnessPresent': {'Yes': 1, 'No': 0},
    'AgentType': {'External': 1, 'Internal': 0}
}

for col, mapping in binary_cols.items():
    x_train[col] = x_train[col].map(mapping)
    x_test[col] = x_test[col].map(mapping)

x_train[['Sex', 'AccidentArea', 'Fault', 'PoliceReportFiled', 'WitnessPresent', 'AgentType']].head()
```

```python
# Map ordinal categorical columns to sequential integers
ordinal_mappings = {
    'VehiclePrice': {
        'less than 20000': 0, '20000 to 29000': 1, '30000 to 39000': 2,
        '40000 to 59000': 3, '60000 to 69000': 4, 'more than 69000': 5
    },
    'AgeOfVehicle': {
        'new': 0, '2 years': 1, '3 years': 2, '4 years': 3,
        '5 years': 4, '6 years': 5, '7 years': 6, 'more than 7': 7
    },
    'AgeOfPolicyHolder': {
        '16 to 17': 0, '18 to 20': 1, '21 to 25': 2, '26 to 30': 3,
        '31 to 35': 4, '36 to 40': 5, '41 to 50': 6, '51 to 65': 7, 'over 65': 8
    },
    'Days_Policy_Accident': {
        'none': 0, '1 to 7': 1, '8 to 15': 2, '15 to 30': 3, 'more than 30': 4
    },
    'Days_Policy_Claim': {
        'none': 0, '8 to 15': 1, '15 to 30': 2, 'more than 30': 3
    },
    'PastNumberOfClaims': {
        'none': 0, '1': 1, '2 to 4': 2, 'more than 4': 3
    },
    'NumberOfSuppliments': {
        'none': 0, '1 to 2': 1, '3 to 5': 2, 'more than 5': 3
    },
    'AddressChange_Claim': {
        'no change': 0, 'under 6 months': 1, '1 year': 2,
        '2 to 3 years': 3, '4 to 8 years': 4
    },
    'NumberOfCars': {
        '1 vehicle': 0, '2 vehicles': 1, '3 to 4': 2,
        '5 to 8': 3, 'more than 8': 4
    }
}

for col, mapping in ordinal_mappings.items():
    x_train[col] = x_train[col].map(mapping)
    x_test[col] = x_test[col].map(mapping)

x_train[list(ordinal_mappings.keys())].head()
```

```python
# One-Hot Encode remaining nominal columns
nominal_cols = ['Make', 'PolicyType', 'MaritalStatus', 'VehicleCategory', 'BasePolicy', 
                'Month', 'MonthClaimed', 'DayOfWeek', 'DayOfWeekClaimed']

# Initialize encoder to drop the first category (to avoid collinearity)
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')

# Fit only on training set nominal columns
encoder.fit(x_train[nominal_cols])

# Transform both training and testing set nominal columns
train_encoded = encoder.transform(x_train[nominal_cols])
test_encoded = encoder.transform(x_test[nominal_cols])

# Convert to DataFrames
encoded_cols = encoder.get_feature_names_out(nominal_cols)
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_cols)
test_encoded_df = pd.DataFrame(test_encoded, columns=encoded_cols)

# Drop original nominal columns and join the new one-hot encoded columns
x_train_preprocessed = x_train.drop(columns=nominal_cols).join(train_encoded_df)
x_test_preprocessed = x_test.drop(columns=nominal_cols).join(test_encoded_df)

print("Preprocessed train shape:", x_train_preprocessed.shape)
print("Preprocessed test shape:", x_test_preprocessed.shape)
```

```python
# Train a Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf_model.fit(x_train_preprocessed, y_train)

# Evaluate model predictions
y_pred = rf_model.predict(x_test_preprocessed)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
rf_model.score(x_train_preprocessed, y_train),rf_model.score(x_test_preprocessed,y_test)
```
